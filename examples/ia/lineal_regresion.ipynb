```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
x = np.array(boston.data[:,5])
y = np.array(boston.target)
```

```python
plt.scatter(x, y, alpha=0.3)
plt.show()
```

```python
# AÃ±adimos columna de 1s para termino independiente
size = x.shape[0]
x_data = np.array([np.ones(size),x]).T
```

```python
# Regresion lineal
r = np.linalg.inv(x_data.T @ x_data) @ x_data.T @ y
# Considerando la line y = mx + b
b = r[0]
m = r[1]
```

```python
# Data de la linea
min_x = x.min()
max_x = x.max()
x_line  = np.array([min_x, max_x])
y_line = m * x_line + b
```

```python
plt.scatter(x, y, alpha=0.3)
plt.plot(x_line, y_line, c="red")
plt.show()
```
