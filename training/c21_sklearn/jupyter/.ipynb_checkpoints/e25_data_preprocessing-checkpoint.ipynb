```python
import sklearn as skl
import pandas as pd
```

```python
# Loading Data Bunch from Scy Kit Learn Package
databunch = skl.datasets.load_boston()
databunch.keys()
```

```python
# Loading Features from Data Bunch
features = pd.DataFrame(databunch.data, columns=databunch.feature_names)
features.head()
```

```python
# Loading Target from Data Bunch
target = pd.DataFrame(databunch.target, columns=['PRICE'])
target.head()
```

```python
# Merging Features and Target
dataframe = pd.DataFrame(features)
dataframe[target.columns] = target
dataframe.head()
```

```python
# Saving Dataframe in a file
outfile = '../data/boston.csv'
dataframe.to_csv(outfile, sep=",", index=False, mode='w')
```
