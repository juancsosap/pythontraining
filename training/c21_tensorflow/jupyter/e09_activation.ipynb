```python
import warnings
warnings.filterwarnings('ignore',category=FutureWarning)

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Defining Result Plotter
def plot(x, y):
    plt.plot(x, y)
    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
# Defining Tensors Runner
def run(sess, xt, yt):
    x, y = sess.run((xt, yt))
    plot(x, y)
```

```python
# create a session to run the tensors
sess = tf.Session()
```

```python
# Defining Tensors
xt = tf.constant(np.linspace(-10, 10, 100), tf.float32)
```

```python
# Linear or Identity Activation Function
def identity(xt):
    return xt

yt = identity(xt) 
run(sess, xt, yt)
```

```python
# Unit or Binary State Activation Function
def binary(xt, threshold):
    ones = tf.ones(tf.shape(xt), tf.float32)
    cond = tf.less(xt, ones * threshold)
    return tf.where(cond, ones * 0, ones)

yt = binary(xt, 5)
run(sess, xt, yt)
```

```python
# Sigmoid or Logistic Activation Function
def sigmoid(xt):
    return 1 / (1 + tf.exp(-xt))

# yt = tf.sigmoid(xt)
yt = sigmoid(xt)
run(sess, xt, yt)
```

```python
# Sigmoid or Logistic Activation Function
def tanh(xt):
    return (tf.exp(xt) - tf.exp(-xt)) / (tf.exp(xt) + tf.exp(-xt)) 

# yt = tf.tanh(xt)
yt = tanh(xt)
run(sess, xt, yt)
```

```python
# Rectified Linear Unit Activation Function
def relu(xt):
    zeros = tf.zeros(tf.shape(xt), tf.float32)
    cond = tf.less(xt, zeros)
    return tf.where(cond, zeros, xt)

# yt = tf.nn.relu(xt)
yt = relu(xt)
run(sess, xt, yt)
```

```python
# Softmax Activation Function
def softmax(xt):
    exps = tf.exp(xt)
    total = tf.reduce_sum(exps)
    return tf.divide(exps, total)

# yt = tf.nn.softmax(xt)
yt = softmax(xt)
run(sess, xt, yt)

total = tf.reduce_sum(yt)
print(sess.run(total))
```

```python
# closing session
sess.close()
```
